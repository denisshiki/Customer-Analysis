<h1>ðŸ§¹ Data Treatment </h1>
<p>In this section we'll clean any missing data and we'll also remove outliers from the attribute "Income".</p>

```python
def data_cleaning(data):
    
    # Change the type of the column to datetime
    data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'])
    
    # Fill the missing data with a 0
    data = data.fillna(0)
    
    # Change some atributes
    data['Year_Birth'] = data['Year_Birth'].astype('int64')
    data['Age'] = data['Year_Birth'].apply(lambda x: int(pd.to_datetime('today').strftime('%Y')) - x)
    data['Age'] = data['Age'].apply(lambda x: 'Teen' if x <= 19 else 
                                   'Adult' if (x > 19) & (x <= 39) else 
                                   'Middle Age Adult' if (x > 39) & (x <= 59) else
                                   'Senior Adult') 
    
    # Drop some data
    data = data.drop(['AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','AcceptedCmp1','AcceptedCmp2',
                'Complain','Z_CostContact','Z_Revenue','Response'], axis = 1)
     
    # Remove Outliers
    value = data['Income']
    Q1 = value.quantile(0.25)
    Q3 = value.quantile(0.75)
    IIQ = Q3 - Q1
    lim_inf = Q1 - IIQ*1.5
    lim_sup = Q1 + IIQ*1.5

    selection = (value >= lim_inf) & (value <= lim_sup)

    new_data = data[selection]
    return new_data
```

```python
new_data = data_cleaning(data)
new_data
```

<h1>ðŸ‘©â€ðŸ’» Data Pre-processing </h1> 

Before we use some Machine Learning classification models we have to make some data preparations following the next steps:  
<ol>
<li>Separate the data atributes between categorical and numerical.</li>
<li>Transform the categorical data into numbers using the library "One Hot Encoder".</li>
<li>Concatenate the both separate data.</li>
<li>Transform the sampled data in a number between 0 and 1.</li>
</ol>

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
def data_processing(data):
    # Get data and drop some columns
    data = data_cleaning(data)
     
    # Instanciating the libraries
    scaler = StandardScaler()
    onehotencoder = OneHotEncoder()
    
    # Define the random seed
    np.random.seed(42)
    
    # Select and processing the categorical data
    select_cat_data = new_data.select_dtypes('object')
    cat_data = onehotencoder.fit_transform(select_cat_data).toarray()
    cat_data = pd.DataFrame(cat_data)
    
    # Select the numerical data
    num_data = new_data.select_dtypes(['int64', 'float64'])
    
    # Concatenating the data
    mark_data = pd.concat([cat_data.reset_index(drop=True), num_data.reset_index(drop=True)], axis = 1)

    # Fit and transform the data
    mark_data_escalar =  scaler.fit_transform(mark_data)
    
    return mark_data_escalar
```

<h1>ðŸ’ª Elbow Method </h1> 

To optimize the classification algorithm and define the number of clusters we use the elbow method  

```python
from sklearn.cluster import KMeans
```

```python
list1 = []
mark_data = data_processing(data)

# Plot a number of clusters between 1 and 20
for i in range(1,20):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(mark_data)
    list1.append(kmeans.inertia_)

plt.figure(figsize=(14,7));
sns.lineplot(data = list1, markers = True);
```

Using this method we see that 4 clusters will give us the best optimization to our algorithm.

<h2> K-Means Model</h2>

First we'll apply the K-Means model and analyse the created clusters to take some conclusions

```python
def kmeans(data): 
    
    # Call the functions needed
    new_data = data_cleaning(data)
    mark_data = data_processing(data)
    
    # Fit the data and return the labels
    kmeans = KMeans(n_clusters = 4)
    kmeans.fit(mark_data)
    labels = kmeans.labels_
    
    # Concat the labels to the treated data
    cluster_data = pd.concat([new_data.reset_index(drop=True), 
                              pd.DataFrame({'Clusters': labels}).reset_index(drop=True)], axis = 1)
    return cluster_data
```

```python
kmeans(data)
```

```python
kmeans_data = kmeans(data)
kmeans_data['Clusters'] = kmeans_data['Clusters'].apply(lambda x: 'hight consumers' if x == 0 else
                                                       'elite consumers' if x == 1 else
                                                       'average consumers' if x == 2 else
                                                       'lower consumers')

#This is a dinamic graphic that will only work when downloaded
#x_selector = widgets.Dropdown(
#    options = kmeans_data.columns ,
#    value= kmeans_data.columns[0],
#    description='X atribute:',
#    disabled=False)

#y_selector = widgets.Dropdown(
#    options = kmeans_data.columns ,
#    value= kmeans_data.columns[0],
#    description='Y atribute:',
#    disabled=False)

#def graphic (kmeans_data, x_selector, y_selector):
#    fig = px.box(kmeans_data, x = x_selector, y = y_selector, color = kmeans_data['Clusters']) 
#    fig.show()
    
#widgets.interactive(graphic, kmeans_data = fixed(kmeans_data), x_selector = x_selector, y_selector = y_selector)
```
